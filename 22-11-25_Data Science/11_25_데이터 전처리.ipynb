## 데이터 프레임 다루기


```python
from patsy import demo_data

# 랜덤값은 고정됨. 표준정규분포에서 Sampling하기 때문에.
df = pd.DataFrame(demo_data('x1','x2','x3','x4','x5'))
df
```

```python
from patsy import dmatrix

# 데이터 여러개 가져올 수 있다.
dmatrix('x1 + x2 + x3 + 0', data=df) # +0(-1)은 상수를 넣지 않겠다.
```

```python
dmatrix('x1 + x2 + x3', data=df)
```

```python
# 넘파이 타입이 아니어도 함수를 가져올 수 있다.
dmatrix('x1  + np.log(np.abs(x2))', data=df)
```

```python
def five_times(x):
  return 5*x

dmatrix('five_times(x1)', data=df)
```

```python
a = lambda x : 5*x
dmatrix('a(x1)', data=df)
```

```python
dmatrix('x1 + x2 + x1:x2 + 0', data=df)
```

```python
dmatrix('x1 + x2 + x1*x2 + 0', data=df)
```

```python
# x3 + x4 값을 받고싶은데 안된다.
dmatrix('x1 + x2 + (x3+x4) + 0', data=df) 
```

```python
# 대문자 I 를 사용하면 된다.
dmatrix('x1 + x2 + I(x1+x2) + 0', data=df)
```

```python
'''
center : 평균을 0으로 scailing.
standardize : 평균을 0으로 하고 표준편차를 1로 scailing.
'''

dmatrix('center(x1)+0', data=df)
```

```python
dmatrix('x1 + center(x1)', data=df)
```

```python
dmatrix('x1 + standardize(x1)', data=df)
```

## sklearn을 통해 데이터 표준화 및 정규화 하기

```python
X = np.arange(7).reshape(-1,1)
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit_transform(X)

# 평균을 빼주는 이유는 이상치가 극단적이라면 데이터 분석이 원만하지 않다. 안드로메다로 간 값을 지구로 들어오게하자.
```

```python
X2 = np.vstack([X, [[10000]]])
X2
```

```python
scaler.fit_transform(X2) # 대칭적으로 이루어지고 있지 않다. 이상치가 있다.
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
scaler.fit_transform(X2)
```

```python
# 다항식으로 만들어 준다.
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3, include_bias=False) # 1은 자기 자신 값, include_bias=False : 상수 x
poly.fit_transform(X)
```

```python
# 본인이 만든 함수(사용자 지정)도 데이터를 변형 할 수 있다.
from sklearn.preprocessing import FunctionTransformer

X = 90*np.arange(9).reshape(-1, 1) # 삼각함수
X
```

```python
def sincos(X):
  x0 = np.sin(X*np.pi/180)
  x1 = np.cos(X*np.pi/180)
  X_new = np.hstack([x0, x1])
  return X_new

X2 = FunctionTransformer(sincos).fit_transform(X)       # 180도 마다 0이 나와야한다.
```

```python
X2 # 스무딩 : 딱 0 값이 나오지 않게 하는 옵션인데 이것이 적용이 된 것인지 아닌지는 모름.
```

```python
plt.subplot(211)
plt.plot(X)
plt.yticks(X.flatten())
plt.title('Degree')
plt.grid()
plt.subplot(212)
plt.plot(X2)
plt.title('Sin Cos')
plt.tight_layout()
plt.grid()
plt.show()

# 우리는 grid와 그래프가 만나는 점(?) 만 데이터로 받는다. 그래서 곡선 처리여도 상관없다.
```
