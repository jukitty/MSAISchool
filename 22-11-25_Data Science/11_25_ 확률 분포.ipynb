# 확률분포의 종류

- 베르누이 분포
      bernoulli
      베르누이 실험 : 결과가 두가지 밖에 없는 실험 ex.동전(앞,뒤)
      베르누이의 확률변수 : 이산확률변수 ex. {0,1} {-1,1} 로 표현
      [여기에서의 뮤는 1이 나올 확률. (보통의 뮤는 평균값을 의미)]
$$Bern(x;μ)={μ^x}{(1-μ)^{(1-x)}}$$ 
- 이항분포
      binom
- 다항분포
      multinomial
- 정규분포
      norm
- 균일분포
      uniform
- 베타분포
      beta
- 감마분포
      gamma
- t - 분포
      t
- 카이제곱분포
      chi2

위 분포를 불러오기 위해선 모수를 지정해줘야 한다.
모수 : 대부분 평균(분포의 기댓값), 편차(표준편자)를 준다.


      pdf : 확률밀도함수 (정규분포를 사용할 때 사용)
      pmf : 확률질량함수 (이산데이터를 사용할 때 사용)
      cdf : 누적확률밀도함수
      ppf : cdf의 역함수

```python
!pip install koreanize-matplotlib
import koreanize_matplotlib

import scipy as sp
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#정규분포 불러오기
rv = sp.stats.norm(1, 2) # 평균, 표준편차 / 표준정규분포:(0,1)

x = np.linspace(-10, 10, 200)
pdf = rv.pdf(x)
plt.plot(x, pdf)
plt.title('확률밀도함수')
plt.grid()
plt.show()

# cdf(누적)를 미분해서 이 그래프를 만들었다. ex. f(-2.5) 일때 선의 넓이는 0이지만 이 모든 넒이를 다 합치면 1이 된다.
```

```python
cdf = rv.cdf(x)
plt.plot(x, cdf)
plt.title('누적확률밀도함수')  # 누적확률밀도함수니까 모든 표본들이 일어날 확률은 1
plt.grid()
plt.show()
```

```python
# 표본 만들기
# rvs : random value sampling
sns.distplot(rv.rvs(size=10000, random_state=0))
plt.title('랜덤 표본 그래프')
plt.xlabel('표본값')
plt.ylabel('count')
plt.grid()
plt.show()

# 모집단 : 평균1, 표준편차(std)2 를 따른 정규분포
# 모집단의 부분집합인 표본집단의 평균과 분산을 사용해서 표준편차를 그린 그래프(모수의 평균과 표준편차와 같다는 전제하에. 원래는 모수의 평균과 표준편차를 그리는 그래프)
```

변환 확률 변수의 시뮬레이션

```python
rv1 = sp.stats.uniform()
rv2 = sp.stats.uniform()

np.random.seed(0)
N = 50000
x_1 = rv1.rvs(N)
x_2 = rv2.rvs(N)

x_3 = x_1 + x_2

plt.subplot(131)
sns.distplot(x_1, kde=False)
plt.grid()
plt.title('균일분포')
plt.xlabel('표본값')

plt.subplot(132)
sns.distplot(x_2, kde=False)
plt.grid()
plt.title('균일분포')
plt.xlabel('표본값')

plt.subplot(133)
sns.distplot(x_3, kde=False)
plt.title('균일분포 표본의 합의 분포')
plt.grid()
plt.tight_layout()
plt.show()

# 균일 분포가 동일하다고 해서 값이 같은 것은 아니다.
```

### 1. 베르누이 분포
      bernoulli
      베르누이 실험 : 결과가 두가지 밖에 없는 실험 ex.동전(앞,뒤)
      베르누이의 확률변수 : 이산확률변수 ex. {0,1} {-1,1} 로 표현
      [여기에서의 뮤는 1이 나올 확률. (보통의 뮤는 평균값을 의미)]
$$Bern(x;μ)={μ^x}{(1-μ)^{(1-x)}}$$ 

```python
mu = 0.6
rv = scipy.stats.bernoulli(mu)
rv
```

```python
x = [0, 1] # 동전의 뒷면0 , 앞면1
plt.bar(x, rv.pmf(x))
plt.xticks([0,1], ['x=0','x=1'])
plt.grid()
plt.xlabel('표본값')
plt.title('베르누이 확률질량 함수')
plt.show()
```

```python
x = rv.rvs(100, random_state=0)
x
```

```python
sns.countplot(x)
plt.title('베르누이의 확률변수 시뮬레이션')
plt.xlabel('표본값')
plt.grid()
plt.show()
```

```python
y = np.bincount(x, minlength=2)/float(len(x))
df = pd.DataFrame({'이론':rv.pmf([0,1]), '시뮬레이션' : y})
df.index = [0,1]
df
```

```python
for i in range(200):
  x = rv.rvs(i, random_state=0)
  y = np.bincount(x, minlength=2)/float(len(x))
  print(f'표본 수가 {i}개 일때 이론값 : {rv.pmf([0,1])}, 표본값 : {y}')
```

###이항분포
베르누이 분포에서 실험을 여러번 한 것이 이항분포   
(n번을 실행하는 베르누이 분포라고 생각하자.)   
X : 성공횟수   
N : 실험 중 시행 수       
      ex. 1번 실험 : 동전 10번, 앞면나올 확률0.6 일때 -> X=6 정도. N=10   

$$ X∼Bin(x;N,μ)$$


표본 중 성공일 경우 1, 아닐경우 0.

N개의 표본 중 성공 횟수는 N개의 표본을 모두 더한 것과 같다.
$$x=\Sigma{y_i}$$

확률 질량 함수
$$Bin(x;N,μ) = \binom{N}{x}{μ^x}{(1-μ)^{(N-x)}}$$

```python
N = 10
mu = 0.6
rv = scipy.stats.binom(N, mu)
```

```python
x = np.arange(N + 1)

plt.bar(x, rv.pmf(x), align='center')
plt.title('이항분포의 확률질량함수')
plt.xlabel('표본값')
plt.ylabel('P(X)')
plt.grid()
plt.show()

# 6회 했을때 1이 나올 확률이 제일 높다.
```

```python
# 1번 실험당 10번씩 던짐. 0번째 실험에서 10번 중 6번 성공. 
np.random.seed(0)
x = rv.rvs(100)
x
```

```python
sns.countplot(x)
plt.title('이항분포의 시뮬레이션 결과')
plt.xlabel("표본값")
plt.show()
```

```python
x_ = np.arange(N + 1)

y = np.bincount(x, minlength=N+1)/float(len(x))
df = pd.DataFrame({'이론':rv.pmf(x_), '시뮬레이션':y}).stack()
```

```python
df = df.reset_index()
df.columns = ['표본값','유형','비율']
```

```python
df
```

```python
sns.barplot(x='표본값', y='비율', hue='유형', data=df)
plt.title('이항분포의 이론적 분포와 표본의 분포')
plt.show()
```

### 정규분포
가우스 분포라고도 함
보통 자연현상에 나타나는 수치 정규분포 가정을 많이 함

$$ N(x;μ,{q^2}) = \frac{1}{{(2π{q^2})}^\frac{1}{2}}exp(-\frac{{(x-μ)}^2}{2{q^2}})$$

```python
mu = 0
std = 1
rv = scipy.stats.norm(mu, std)
```

## Q-Q 플롯

어떤 확률변수의 분포가 정규분포인지 검사하는 그래프

정규분포를 따를 때 Q-Q플롯은 직선의 형태를 가집니다. 

정규분포를 따르지 않을 때 Q-Q플롯은 곡선이 발생

```python
mu = 0
std = 1
rv = scipy.stats.norm(mu, std)
```

```python
x = rv.rvs(100)
x_sorted = np.sort(x)
x_sorted
```

```python
from scipy.stats.morestats import _calc_uniform_order_statistic_medians

position = _calc_uniform_order_statistic_medians(len(x))
position
```

```python
ff = rv.ppf(position)
ff
```

```python
# 1자로 된 형식이 보인다. 이런 모양이 정규분포를 따른다고 얘기한다.
# x가 커질수록, 많이 뽑을수록 1자가 된다.
plt.scatter(ff, x_sorted)
plt.title('QQ')
plt.axis('equal')
plt.show()
```

### 중심 극한 정리

여러개의 확률변수의 합이 정규분포와 비슷한 분포를 이루는 현상

확률변수의 평균들의 평균은 확률변수의 수가 무한히 많아지게 되면 모집단의 평균과 분산이 된다.

```python
np.random.seed(0)
xx = np.linspace(-2,2,100)

plt.figure(figsize=(6,9))

for i, N in enumerate([1,2,10, 100]):
  X = np.random.rand(5000, N)
  Xbar = (X.mean(axis=1)-0.5)*np.sqrt(12*N)
  ax = plt.subplot(4,2,2*i+1)
  sns.distplot(Xbar, bins=10, kde=False, norm_hist=True)
  ax.set_title(f'N = {N}')
  plt.subplot(4,2,2*i+2)
  sp.stats.probplot(Xbar, plot=plt)

plt.tight_layout()
plt.show()
```

### 스튜던트 t분포

정규분포에서 양끝단의 분포가 많아지는 분포를 t분포(fat tail 현상)

```python
# import panads_datareader.data as web
import pandas_datareader.data as web
symbols = ['SP500','NASDAQCOM','DJCA', 'NIKKEI225']
data = pd.DataFrame()
for sym in symbols:
  data[sym] = web.DataReader(sym, data_source='fred')[sym]
data = data.dropna()
```

```python
(data/data.iloc[0]*100).plot()
plt.xlabel('날짜')
plt.ylabel('주가 수익율')
plt.grid()
plt.show()
```

```python
log_returns = np.log(data/data.shift(1))
log_returns.hist(bins=50)
plt.tight_layout()
plt.show()
```

```python
for i, sym in enumerate(symbols):
  ax = plt.subplot(2, 2, i+1)
  sp.stats.probplot(log_returns[sym].dropna(),plot=ax)
plt.tight_layout()
plt.show()
```

```python
xx = np.linspace(-4,4,100)
for df in [1, 2, 5, 10, 20]:
  rv = sp.stats.t(df=df)
  plt.plot(xx, rv.pdf(xx), label=f'스튜던트 t 분포(df={df})')
plt.plot(xx, sp.stats.norm().pdf(xx), label='정규분포', lw=5, alpha=0.5)
plt.title('자유도에 따른 스튜던트 t분포의 변화')
plt.ylabel('P(x)')
plt.legend()
plt.show()
```

### 카이제곱 분포

t분포와는 다르게 제곱을해서 양수값만 가지도록 만든것이 카이제곱 분포


```python
xx = np.linspace(0.01, 10, 100)
dfs = np.arange(1,7)
lss = ['-','--','-.',':','-.','-.']

for df, ls in zip(dfs, lss):
  rv = scipy.stats.chi2(df=df)
  plt.plot(xx, rv.pdf(xx), ls=ls, label=f"자유도 : {df}")
plt.title('자유도에 따른 카이제곱 분포')
plt.ylim(0,0.6)
plt.xlabel('표본값')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```

### F-분포
여러 집단에 대해서 각 집단의 평균이 의미있게 차이가 나는지 측정 할 때 사용됩니다.   
선형 회귀 분석에 회귀분석모델이 전체 데이터 변동을 잘 설명하고 있는지 비교하기 위해서 사용됩니다

```python
xx = np.linspace(0.03, 3, 1000)
plt.plot(xx, scipy.stats.f(1,2).pdf(xx), ls = '-', label='F(1,2)')
plt.plot(xx, scipy.stats.f(3,5).pdf(xx), ls = '-', label='F(1,2)')
plt.plot(xx, scipy.stats.f(10,10).pdf(xx), ls = '-', label='F(1,2)')
plt.title('자유도에 따른 F분포의 모양')
plt.legend()
plt.show()
```

# 확률분포의 결정
확률변수가 베르누이, 이항, 정규분포 중에 어떤 확률분포를 따르는지 알아내고, 표본집단의 해당 확률분포의 모수 값을 구하는 작업


확률 변수가 어떤 분포를 따르는지 데이터를 보면 추측가능.
(데이터의 특성을 알면 어떤 분포를 따를지 추측이 가능하다.)

      - 데이터가 0, 1 뿐일 경우 : 베르누이 분포
      - 데이터가 카테고리 값일 경우: 카테고리 분포
      - 테이터가 0과 1사이의 실수 값일 경우 : 베타분포
      - 데이터가 항상 0 이상일 경우 : 로그정규분포, 감마분포, f분포, 카이제곱분포, 지수 등등
      - 데이터가 제한이 없고 연속된 실수 일 경우 : 정규분포, 스튜던트 t-분포, 라플라스 분포
