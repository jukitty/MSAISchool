# 파이차트

```python
ratio = [34, 32, 16, 18] # 차지하는 비율
labels = ['Apple', 'Banana', 'Melon', 'Grapes'] # 라벨 이름
explode = [0.05, 0.05, 0.05, 0.05] # 중심에서 벗어나는 비율
colors = ['silver', 'gold', 'whitesmoke', 'lightgray'] # 색상

# 파이차트를 그리는 메서드 plt.pie
plt.pie(ratio, labels=labels, 
        autopct ='%.1f%%',
        startangle=260,      # 260도에서 시작한다.
        counterclock=False,
        explode=explode,
        shadow=True,
        colors=colors
 )
plt.show()
```

```python
 wedgeprops={'width': 0.7, 'linewidth': 5}

plt.pie(ratio, labels=labels, 
        autopct ='%.1f%%',
        startangle=0,
        counterclock=True,
        colors=colors,
        wedgeprops=wedgeprops
)
```

# 밀도 그래프

```python
import numpy as np
from scipy import stats
```

```python
# 정규분포 만들기 normal
# 중앙값이 2이고 표준편차가 2인 랜덤 데이터 1000개 생성
data_a = np.random.normal(2.0, 2.0, 100)
data_a
```

```python
plt.plot(data_a)
plt.show()
```

```python
estimator = stats.gaussian_kde(data_a, bw_method='silverman') # 다음에 배울 내용
x = np.arange(-10,10, 0.01)
K = estimator(x)
```

```python
plt.plot(x, K, label='gaussian_kde')
```

# 박스 그래프

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.seed(0)

plt.style.use('default')

data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

fig, ax = plt.subplots()

# 박스 그래프 : boxplot 호출 
ax.boxplot([data_a, data_b, data_c], notch=True, whis=1.5, vert=True )
ax.set_ylim(-10.0, 10.0) # y에 최저값과 최대값을 표시하기 위해
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
plt.show()
```

# Violin plot

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')

# 데이터를 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 바이올린 : plt.violinplot()
violin = plt.violinplot([data_a,data_b,data_c],
                        showmeans=True,
                        showextrema=True,
                        showmedians=True # 중앙값 : 표시 데이터를 내림이나 오름차순 했을 때 중앙에 있는 값
)
plt.xticks(np.arange(1,4), labels=['A','B','C'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')

# 데이터를 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 바이올린 : plt.violinplot()
violin = plt.violinplot([data_a,data_b,data_c],
                        showmeans=True,
                        showextrema=True,
                        showmedians=True # 중앙값 : 표시 데이터를 내림이나 오름차순 했을 때 중앙에 있는 값
)
plt.xticks(np.arange(1,4), labels=['A','B','C'])
violin['bodies'][0].set_facecolor('red')
violin['cbars'].set_edgecolor('red')
violin['cmaxes'].set_edgecolor('orange')
violin['cmins'].set_edgecolor('green')
violin['cmeans'].set_edgecolor('red')
violin['cmedians'].set_edgecolor('red')
plt.show()
```
