```python
fig = plt.figure()    # 그림판을 만들었지만 속이 비어있다
ax =  plt.axes()      # 빈 축을 만들어두자

x = np.linspace(0,10,100)
ax.plot(x, np.sin(x))       # 빈 축에 선을 그리자
```

```python
             # sin(x-1) : y = -1 시작
plt.plot(x, np.sin(x-0), color='blue')          # 색상의 이름을 지정
plt.plot(x, np.sin(x-1), color='g')             # 짧은 색상 코드. RGB, CMYK
plt.plot(x, np.sin(x-2), color='0.75')          # 0-1 사이 회색조 표현. 빛 밝기.
plt.plot(x, np.sin(x-3), color='#FFDD44')       # 16진수로 표현한 RGB
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3)) # RGB값을 튜플로 표현
plt.plot(x, np.sin(x-5), color='blue')
```

```python
# 선 Style 지정

plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r')
```

```python
# 플롯 조정하기 : 축 경계

plt.plot(x, np.sin(x))

plt.xlim(-1,11)     # x 축의 limit
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)    # 시작 점 조정
plt.ylim(1.2,-1.2)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')  # tight, equal
```

```python
# 플롯에 레이블 붙이기
# 데이터가 많아졌을때 꼭 사용

plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-g',label='sin(x')      # 선에 라벨링을 했을 뿐 나오지 않음
plt.plot(x, np.cos(x), ':b',label='sin(x')
plt.axis('equal')

plt.legend()      # 라벨은 이 때 사용
```

```python
# 산점도(점으로 표현)의 출력

x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')  #'o' 자리를 마커라고 함. 표시하는 방법.
```

```python
rng = np.random.RandomState(0)  # NumPy 메서드로 랜덤 값을 만들어줌. 0번 시드는 값 고정.
for marker in ['o',',','x','+','v','<','>','s','d']:
  plt.plot(rng.rand(5), rng.rand(5), marker,
           label='marker={0}'.format(marker)) # 좌표 랜덤 값. marker 순차적용
  plt.legend()
  plt.xlim(0,1.8)
```

```python
plt.plot(x,y,'-ok') # o 는 marker
```

```python
plt.plot(x, y, '-p', color='gray', 
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='red',
         markeredgewidth='2')
```

```python
plt.scatter(x, y, marker='o')      # ex) 인구수에 따라 원 크기차이
```

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
color = rng.randn(100)
size = 1000 * rng.randn(100) # 실제로 그리는 코드는 한줄

plt.scatter(x,y,s=size,c=color,alpha=0.3,cmap='magma') # s:사이즈, alpha:투명도, colorbar: 범주(색마다)
plt.colorbar()
```

```python
# sckit-learn = 머신러닝 framwork (딥러닝 사용x)
from sklearn.datasets import load_iris    # 아이리스 꽃 데이터 가져오기

iris = load_iris()
```

```python
# 어떻게 생겼는지 꼭 확인
#iris.shape -> 에러, 넘파이 배열x
type(iris)
iris.data # sklearn에서 지원하는 검색. 배열은 나오지만 무엇에 해당되는 건지 모름.
```

```python
iris.data.T # 전체 데이터의 배열
```

```python
iris.target  # 총 150개 0,1,2 아이리스 꽃은 3가지의 종류
```

```python
features = iris.data.T
```

```python
plt.scatter(features[0], features[1], s=100*features[3],c=iris.target, alpha=0.3,cmap='magma') # 길이와 폭에 해당하는 데이터
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
# 오차 시각화
x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')    #  yerr:오차범위 fmt:출력방식
```

```python
plt.errorbar(x, y,yerr=dy,fmt='o',color='black',
             ecolor='lightgray', elinewidth=3, capsize=0)
```
