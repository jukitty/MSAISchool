```python
# 데이터 준비 과정

from sklearn.model_selection import train_test_split  # 데이터 쪼개기. 학습용&테스트용
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0, test_size=0.3) 
                                                    # 랜덤값을 주면 알아서 섞은 후 각 라벨 4가지의 값, 테스트용 사이즈 30% 비율로 잘라준다. (7:3 비율)
```

```python
print('X_train의 크기:', X_train.shape)
print('y_train의 크기:', y_train.shape)
```

```python
print('X_test의 크기:', X_test.shape)
print('y_test의 크기:', y_test.shape)
```

```python
# 데이터의 시각화  <dataframe객체로>

import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset['feature_names'])
```

```python
pd.plotting.scatter_matrix(iris_dataframe, figsize=(15,15), 
                           marker='o', 
                           c=y_train,
                           cmap='magma',
                           alpha=0.8)
```

```python
# 첫 번째 머신 러닝 모델: k-최근접 알고리듬

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)   # 이웃의 갯수는 홀수가 좋지만 케바케.
```

```python
knn.fit(X_train, y_train) # 데이터, 라벨값
```

```python
# 예측하기
import numpy as np

X_new = np.array([[5,2.9,1,0.2]]) # 예측용 샘플 2차원 데이터
X_new   # 2차원 데이터 안에 1개의 항목. 항목 들어간 갯 수 만큼 예측
```

```python
knn.predict(X_new)  # setosa 로 예측
```

```python
prediction = knn.predict(X_new)
print('예측:', prediction)
print('예측한 target의 이름:', iris_dataset['target_names'][prediction])
```

```python
# 모델 평가 방법

y_pred = knn.predict(X_test) 
print('테스트 세트에 대한 예측 값:\n', y_pred)
```

```python
y_pred==y_test  # y_test 실제값 & y_pred 예측값 비교
```

```python
print('테스트 세트에 대한 정확도:\n', np.mean(y_pred==y_test)*100)  # 전체의 평균 -> 97%의 정확도를 갖는 모델이다.
```

```python
print('테스트 세트에 대한 정확도:\n', knn.score(X_test, y_test))
```
