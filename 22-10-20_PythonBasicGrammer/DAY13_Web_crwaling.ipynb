```python
import bs4        # html 구조 파악
import requests   # 실제 통신
```

```python
url = 'https://search.shopping.naver.com/search/category/100005307'
result = requests.get(url).text     # 이미지:content, html:text
result
```

```python
bsObj = bs4.BeautifulSoup(result, 'html.parser')   # html을 가장 잘 파싱. 파서:내용 파싱(이해)하는 것
```

```python
items = bsObj.find_all('a',{'class':'basicList_link__JLQJf'}) # a 태그에서 class항목 찾고 text찾기
```

```python
print(items[0].text)
print(items[1].text)
print(items[2].text)
print(items[3].text)
```

```python
lists = items[:3]
for l in lists:
  print(l.text)
```
