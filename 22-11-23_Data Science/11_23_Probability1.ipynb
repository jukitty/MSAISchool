#확률
S = 표본 공간

- 사건이 일어날 확률

- 고전적 확률 정의
```
  - 실험 결과가 N가지일 때 일어난 사건 A가 m번 일어났으면 P(A) = m/N

  - 각 실험 가능성은 모두 같을 때

  - ex. 주사위[1,2,3,4,5,6] 한번 던졌을때 각 각 나올 확률 1/6
```
- 경험적 정의
```
  - 반복 횟수를 N이라 했을 때 무한번 반복했을 때 발생하는 A의 횟수 : m

  - P(A) = lim m/N

  - 주사위를 한번 던지는 것이 아닌 많이 던져서 나온 확률. 고전적 확률과는 다른 
결과.

  - 공정한 주사위라면 무한정 던졌을 때 고전적 확률에 준한 결과가 나올 것.

- 공리적 정의

  - 콜모코로프(사람이름)이 정리한 공리적 정의 세가지   
```
    - 0 <= P(A) <= 1
    - P(S) = 1
    - 실험이 표본공간(S)의 부분집합 이면서 다른 사건과 교집합이 없을 때 각 실험의 확률을 더한 값과 합집합의 확률은 같다.
    - ex. 동전 하나 던지기를 했을 때,
            앞면 : H, 뒷면 : T
            표본공간 = S(HH, HT, TH, TT)
            앞이 안 나오는 사건(A({TT}) 과 뒷면이 안나오는 사건(B{HH})이면 A와 B는 교집합이 없다.
                  - AUB = P({TT,HH}) = P({TT}) + P({HH})    
                  - P(AUB) = P(A) + P(B)



# 확률의 성질
- 공집합의 확률 : 0
- 여집합의 확률   
      실험 A의 여집합은 1-P(A)
- 포함-배제 원리
      P(AUB) = P(A) + P(B) - P(AnB)
- 각 사건이 배반 사상(event)일 경우 (교집합X)
      P(A1 U A2 U A3 ... U An) = P(An)
      A가 B의 부분집합일 때 P(A) <= P(B)

# 조건부 확률
- 사건 B가 발생했을 때(P(B) > 0) 사건 A가 발생할 확률
$$P(A|B) = \frac{P(A\cap B)}{P(B)}$$
- 사건 A가 발생했을 때 사건 B가 발생할 확률    
$$P(B|A) = \frac{P(A\cap B)}{P(A)}$$   
$$P(A\cap B) = P(A|B)P(B) = P(B|A)P(A)$$

- 독립사건
$$P(A|B) = P(A)$$   
$$P(B|A) = P(B)$$


# 베이즈 정리
- 분할사건
      * 조건 2가지*
      1. 사건 K개가 일어났을 때 A1, A2, ... AK 서로 다 교집합이 없는 경우(배반사건) 분할사건
      2. 사건 K개를 모두 합집합했을 때 S가 되는 경우

#### 사전 확률을 사후 확률로 추정하는 정리

$$ B = (B \cap {A_1}) U (B \cap {A_2})U ... U (B \cap{A_k})$$   
$$ P(B) = P(B \cap {A_1} + ... + P(B \cap {A_k})$$   
$$ P({A_i}|B) = \frac{P(B|{A_i})P({A_i})}{P(B)} = \frac {P(B|{A_i})P{B_i})}{P(B|{A_1})P({A_1}) + P(B|{A_2})P({A_2}) + ... + P(B|{A_k})P({A_k})}   $$

#분산과 표준편차
- 분산 = 표준편차의 제곱
- 분산 확률 분포 함수의 모양을 설명할 수 있다.
- 분산은 확률 분포 함수에 확률이 모여있는지 퍼져있는지를 알 수 있는 수치적 파라메터. 

$$ {σ^2} = Var[X] = E[(X-\mu)^2]$$

- 분산의 성질 (확률변수 X에 대한 성질)
      Var[X] >= 0
      Var[cX] = c^2 Var[X]

      Var[X*Y] = Var[X] + Var[Y] + 2E[(X-mx)(Y-my)]   //mx : x의 평균 E : 평균
 
      
- 표본 평균의 분산과 표본 분산의 관계
$$E[X] = E[\bar{X}]$$
$$Var[\bar{X}] = \frac{1}{N}Var[X]$$

 N :  표본집단의 표본 수
